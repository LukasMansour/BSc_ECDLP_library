```python
from api.CircuitChooser import CircuitChooser
from api.NameFilters import custom_name_filter
```

```python
CircuitChooser()._name_filter = custom_name_filter({"BasicConstantAdderIP", "BasicIncrementer"})
CircuitChooser().choose_component("QCAdderIP", (4, 5, 0), dirty_available=0,
                                  clean_available=0).get_circuit().decompose().draw(
    output='mpl')
```

```python
CircuitChooser().clear_caches()
CircuitChooser()._name_filter = custom_name_filter({"BasicConstantAdderIP", "HRSIncrementer"})
# Cheeky case, where you can get away with far less qubits.
# Since adding 6 you get to borrow x_0 (+2 and +4) and you can borrow x_1 (+2).
CircuitChooser().choose_component("QCAdderIP", (4, 6, 0), dirty_available=2, clean_available=0).get_circuit().decompose(
    reps=2).draw(
    output='mpl')
```

```python
CircuitChooser().clear_caches()
CircuitChooser()._name_filter = custom_name_filter({"BasicConstantAdderIP", "HRSCleanIncrementer"})
# Cheeky case, where you can get away with far less qubits.
# Since adding 6 you get to borrow x_0 (+2 and +4) and you can borrow x_1 (+2).
CircuitChooser().choose_component("QCAdderIP", (4, 6, 0), dirty_available=0, clean_available=3).get_circuit().decompose(
    reps=2).draw(
    output='mpl')
```
